```python
import pandas as pd
```

```python
df=pd.read_excel(r'C:\Users\Atiq\Downloads\Pandas Data_cleaning project\Customer Call List.xlsx') 
df
```

<h2><b>Cleaning special symbols from columnn 'Last_Name'</b><h/2>

```python
#one way to remove '...,/,_' is
df['Last_Name']=df['Last_Name'].str.lstrip('...')
df['Last_Name']=df['Last_Name'].str.lstrip('/')
df['Last_Name']=df['Last_Name'].str.rstrip('_')
df
```

```python
# other way to remove '...,/,_' is
df['Last_Name'].str.strip('...,/,_')
df
```

<h1><b>Removing duplicates from the data  </b></h1>

```python
df.drop_duplicates()
```

<h1><b>Removing extra columns from the data  </b></h1>

```python
df=df.drop(columns='Not_Useful_Column')
df
```

<h1><b>keeping standardized phone numbers format in the data  </b></h1>

```python
df["Phone_Number"] = df['Phone_Number'].str.replace(r'\D', '',regex=True) 
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('--', '')
df
```

<h1><b>keeping YES and NO in same format in the data  </b></h1>

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

<h1><b>Replacing and filling null values in the data  </b></h1>

```python
df = df.replace('N/a','')
df = df.fillna('')
df
```

<h1><b>Dropping null values in the data  </b></h1>

```python
df = df.dropna()
df
```

<h1><b>Spliting and Expanding a column in the data  </b></h1>

```python
df[['street', 'state', 'zipcode']] = df['Address'].str.split(', ', expand=True)
df
```

<h1><b>Dropping particular null values within specified column  </b></h1>

```python
for x in df.index :
    if df.loc[x, "Phone_Number"] == '' :
        df.drop(x, inplace=True)

df
```

<h1><b>Resetting new index and dropping the old one  </b></h1>

```python
df = df.reset_index(drop=True)
df
```
